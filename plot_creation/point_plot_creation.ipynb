```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams["figure.figsize"] = (12, 6)
matplotlib.rcParams["font.size"] = 18
```

```python
SVGP_df = pd.read_csv('results/SVGP_data_split.csv')
NN_df = pd.read_csv('results/NN_data_split_patience_7.csv')
Poly_df = pd.read_csv('results/Poly_data_split.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot for Mean Absolute Percentage Error
ax1.plot(SVGP_df['points'], SVGP_df['corrected'], label='SVGP', color = 'blue')
ax1.grid(color='gray', linestyle='--', linewidth=0.5)
ax1.plot(NN_df['points'], NN_df['mape'], label='NN', color='red')
#ax1.plot(Poly_df['points'], Poly_df['mape'], label='Poly', color='green')
ax1.set_xlabel('Training Points')
ax1.set_ylabel('Mean Absolute Percentage Error (%)')
ax1.legend()

# Plot for Time
ax2.plot(SVGP_df['points'], SVGP_df['time'], label='SVGP', color = 'blue')
ax2.grid(color='gray', linestyle='--', linewidth=0.5)
ax2.plot(NN_df['points'], NN_df['time'], label='NN', color='red')
#ax2.plot(Poly_df['points'], Poly_df['time'], label='Poly', color='green')
ax2.set_xlabel('Training Points')
ax2.set_ylabel('Time (s)')
ax2.legend()
ax2.set_axisbelow(True)
ax2.yaxis.grid(color='gray', linestyle='dashed')

fig.tight_layout()
plt.show()
```
